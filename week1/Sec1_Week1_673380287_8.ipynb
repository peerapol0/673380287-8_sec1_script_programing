- Objective: Students will define variables of different data types, perform simple operations, and display results.


2. Declare and print variables:


```python
Age = 21
GPA = 3.99
Favorite_Color = "Blue"
Student = True

print("Age = ",Age ,type(Age))
print("GPA = ",GPA,type(GPA))
print("Favorite_Color = ",Favorite_Color,type(Favorite_Color))
print("Student = ",Student,type(Student))
```

3. Perform simple arithmetic:


```python
num1 = 7
num2 = 6
sum = num1 + num2
print("Sum = ",sum)
```

4. String concatenation and f-strings:


```python
First_name = "Sappy"
Last_name = "Kampy"
Full_name = First_name + " " + Last_name
print(f"Hello,{Full_name}")
```

5. User input and type conversion:


```python
num = input("What your favorite number : ")
print("type of A is : ", type(num))
num = int(num)
result = num + 5
print(num," + 5 = ", result)
```
