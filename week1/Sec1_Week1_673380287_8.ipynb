2. Declare and print variables:


```python
Age = 19
GPA = 3.51
Favorite_Color = "Blue"
student = True

print("Age = ",Age ,type(Age))
print("GPA = ",GPA,type(GPA))
print("Favorite_Color = ",Favorite_Color,type(Favorite_Color))
print("Student = ",student,type(student))
```

3. Perform simple arithmetic:


```python
num1 = 15
num2 = 3
sum = num1 + num2


different = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("num1 = ",num1)
print("num2 = ",num2,)

print("Sum = ",sum)
print("Different = ",different)
print("Product = ",product)
print("Quotient = ",quotient)
```
